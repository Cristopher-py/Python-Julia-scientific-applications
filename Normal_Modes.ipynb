## Two coupled oscillators

<div style = "text-align: center;">
    
</div>

The equations of motion that describe this system are:

\begin{equation*}
    \begin{cases}
    m_1 \, \frac{d ^ 2 x_1 (t)}{d t ^ 2} = - (k_1 + k_2) \, x_1 (t) + k_2 \, x_2 (t) \\
    m_2 \, \frac{d ^ 2 x_2 (t)}{d t ^ 2} = k_2 \, x_1 (t) - (k_2 + k_3) \, x_2 (t)
    \end{cases}
\end{equation*}

One can rewrite this in matrix notation:

\begin{equation*}
\begin{pmatrix}
    m_1 & 0 \\
    0 & m_2 \\
\end{pmatrix}

\,

\frac{d ^ 2}{d t ^ 2}
\begin{pmatrix}
    x_1 (t) \\
    x_2 (t) \\
\end{pmatrix}

= - \,

\begin{pmatrix}
    k_1 + k_2 & - k_2 \\
    - k_2 & k_2 + k_3 \\
\end{pmatrix}

\begin{pmatrix}
    x_1 (t) \\
    x_2 (t) \\
\end{pmatrix}
\end{equation*}

Symbolically:

\begin{equation*}
    M \frac{d ^ 2 \vec{x}}{d t ^ 2} = - K \vec{x}
\end{equation*}

To obtain the basis of all solutions, just substitute:

\begin{equation*}
\begin{pmatrix}
    x_1 (t) \\
    x_2 (t) \\
\end{pmatrix}

=

e ^ {i \omega t}
\begin{pmatrix}
    \alpha _1 \\
    \alpha _2 \\
\end{pmatrix}
\end{equation*}

Or:

\begin{equation*}
    \vec{x} = e ^ {i \omega t} \, \vec{\alpha}
\end{equation*}

Thus we get:

\begin{equation*}
    \left( K - \omega ^ 2 M \right) \vec{\alpha} = 0
\end{equation*}

From the equation above we can compute the $\vec{\alpha}$ vector, while the angular frequencies of normal solutions are provided by solving the equation:

\begin{equation*}
    \left| K - \omega ^ 2 M \right| = 0
\end{equation*}


```python
# Importing libraries
import sympy as smp
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("fast")
```

```python
# Defining symbols
k1, k2, k3, m1, m2 = smp.symbols("k_1, k_2, k_3, m_1, m_2", real = True, positive = True, constant = True)
w = smp.symbols("omega", real = True, positive = True)
```

```python
# Defining the matrices
K = smp.Matrix([[k1 + k2, - k2],
                [- k2, k2 + k3]])

M = smp.Matrix([[m1, 0],
                [0, m2]])
```

```python
K
```

```python
M
```

```python
# Computing A
A = K - w ** 2 * M
A
```

```python
# Computing the determinant of K - w ^ 2 M
ch = A.det().collect(w)
ch
```

```python
# Computing the normal frequencies
omega = smp.solve(ch, w)
omega
```

```python
omega[0]
```

```python
omega[1]
```

Let's consider the simpler case $m_1 = m_2 = m$ and $k_1 = k_2 = k_3 = k$

```python
# Defining symbols
k, m = smp.symbols("k, m", real = True, positive = True, constant = True)
```

```python
# First normal mode
w0 = omega[0].subs([(m1, m), (m2, m), (k1, k), (k2, k), (k3, k)]).simplify()
w0
```

```python
# Second normal mode
w1 = omega[1].subs([(m1, m), (m2, m), (k1, k), (k2, k), (k3, k)]).simplify()
w1
```

```python
# Rewriting A (first normal mode)
A0 = A.subs([(m1, m), (m2, m), (k1, k), (k2, k), (k3, k), (w, w0)])
A0
```

```python
# Computing the alpha vector
alpha0 = A0.nullspace()[0]
alpha0
```

```python
# Rewriting A (second normal mode)
A1 = A.subs([(m1, m), (m2, m), (k1, k), (k2, k), (k3, k), (w, w1)])
A1
```

```python
# Computing the alpha vector
alpha1 = A1.nullspace()[0]
alpha1
```

```python
# Defining the time and some constants
t = smp.symbols("t", real = True, nonnegative = True)
C0, C1 = smp.symbols("C_0, C_1", real = True, positive = True, constant = True)
```

```python
# Computing the general equations of motion
sol = smp.re(alpha0 * C0 * smp.exp(smp.I * w0 * t) + alpha1 * C1 * smp.exp(smp.I * w1 * t))
sol
```

$C_1$ and $C_2$ are provided by the initial conditions of the system

```python
# Going numerical
x1 = smp.lambdify([C0, C1, k, m, t], sol[0], "numpy")
x2 = smp.lambdify([C0, C1, k, m, t], sol[1], "numpy")
```

```python
# Solving the equation of motion for one choice of the parameters
t_num = np.linspace(0, 10, 1000)
C0_num, C1_num, k_num, m_num = (0.5, 0.75, 10, 4)
x1 = x1(C0_num, C1_num, k_num, m_num, t_num)
x2 = x2(C0_num, C1_num, k_num, m_num, t_num)
```

```python
# Creating the plot
plt.figure(figsize = (10, 6))

plt.subplot(1, 2, 1)
plt.plot(t_num, x1, label = r"$x_1$ (t)", color = "red")
plt.plot(t_num, x2, label = r"$x_2$ (t)", color = "blue")
plt.title("Laws of motion", color = "gold", fontweight = "bold")
plt.xlabel("Time t [s]")
plt.ylabel("Displacement [m]")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x1, x2, color = "black", label = "Trajectory")
plt.scatter(x1[0], x2[0], label = "Starting point", color = "green", s = 50)
plt.scatter(x1[-1], x2[-1], label = "End point", color = "red", s = 50)
plt.title("Configuration space trajectory", color = "gold", fontweight = "bold")
plt.xlabel(r"$x_1$ (t) [m]")
plt.ylabel(r"$x_2$ (t) [m]")
plt.legend()

plt.tight_layout()
plt.show()
```

## Three coupled oscillators

For three masses connected by 4 springs, the system of equations would be:

\begin{equation*}
    \begin{cases}
        m_1 \, \frac{d ^ 2 x_1 (t)}{d t ^ 2} = - (k_1 + k_2) \, x_1 (t) + k_2 \, x_2 (t) \\
        m_2 \, \frac{d ^ 2 x_2 (t)}{d t ^ 2} = k_2 \, x_1 (t) - (k_2 + k_3) \, x_2 (t) + k_3 \, x_3(t) \\
        m_3 \, \frac{d ^ 2 x_3 (t)}{d t ^ 2} = k_3 \, x_2 (t) - (k_3 + k_4) \, x_3 (t)
    \end{cases}
\end{equation*}

In matrix form:

\begin{equation*}
\begin{pmatrix}
m_1 & 0 & 0 \\
0 & m_2 & 0 \\
0 & 0 & m_3 \\
\end{pmatrix}

\frac{d ^ 2}{d t ^2}

\begin{pmatrix}
x_1 (t) \\
x_2 (t) \\
x_3 (t) \\
\end{pmatrix}

= -

\begin{pmatrix}
k_1 + k_2 & - k_2 & 0 \\
- k_2 & k_2 + k_3 & - k_3 \\
0 & - k_3 & k_3 + k_4 \\
\end{pmatrix}

\begin{pmatrix}
x_1 (t) \\
x_2 (t) \\
x_3 (t) \\
\end{pmatrix}
\end{equation*}

As seen before, we can calculate the basis of all solutions by substituting in the system the trivial vector:

\begin{equation*}
\begin{pmatrix}
    x_1 (t) \\
    x_2 (t) \\
    x_3 (t)
\end{pmatrix}

=

e ^ {i \omega t}
\begin{pmatrix}
    \alpha _1 \\
    \alpha _2 \\
    \alpha _3
\end{pmatrix}

=

e ^ {i \omega t} \, \vec{\alpha}
\end{equation*}

Getting:

\begin{equation*}
    \left( K - \omega ^ 2 M \right) \vec{\alpha} = 0 \quad
    \left| K - \omega ^ 2 M \right| = 0
\end{equation*}

```python
k4, m3 = smp.symbols("k_4, m_3", real = True, positive = True, constant = True)
```

```python
K = smp.Matrix([[k1 + k2, - k2, 0],
                [- k2, k2 + k3, -k3],
                [0, - k3, k3 + k4]])

M = smp.Matrix([[m1, 0, 0],
                [0, m2, 0],
                [0, 0, m3]])
```

```python
K
```

```python
M
```

```python
A = (K - w ** 2 * M)
A
```

Let's simplify the problem again by setting all $k$ and $m$ equal

```python
# Simplifying the matrix
A = A.subs([(m1, m), (m2, m), (m3, m), (k1, k), (k2, k), (k3, k), (k4, k)])
A
```

```python
ch = A.det().collect(w)
ch
```

```python
omega = smp.solve(ch, w, simplify = True)
omega
```

```python
w0 = omega[0]
w0
```

```python
w1 = omega[1]
w1
```

```python
w2 = omega[2]
w2
```

```python
A0 = A.subs(w, w0)
A0
```

```python
alpha0 = A0.nullspace()[0]
alpha0
```

```python
A1 = A.subs(w, w1)
A1
```

```python
alpha1 = A1.nullspace()[0]
alpha1
```

```python
A2 = A.subs(w, w2)
A2
```

```python
alpha2 = A2.nullspace()[0]
alpha2
```

```python
C2 = smp.symbols("C_2", real = True, positive = True, constant = True)
```

```python
# General solution
sol = smp.re(C0 * alpha0 * smp.exp(smp.I * w0 * t) + C1 * alpha1 * smp.exp(smp.I * w1 * t) + C2 * alpha2 * smp.exp(smp.I * w2 * t))
sol
```

```python
# Going numerical
x1 = smp.lambdify([C0, C1, C2, k, m, t], sol[0], "numpy")
x2 = smp.lambdify([C0, C1, C2, k, m, t], sol[1], "numpy")
x3 = smp.lambdify([C0, C1, C2, k, m, t], sol[2], "numpy")
```

```python
C0_num, C1_num, C2_num, k_num, m_num = (0.5, 0.5, 0.5, 10, 2)
x1 = x1(C0_num, C1_num, C2_num, k_num, m_num, t_num)
x2 = x2(C0_num, C1_num, C2_num, k_num, m_num, t_num)
x3 = x3(C0_num, C1_num, C2_num, k_num, m_num, t_num)
```

```python
plt.figure(figsize = (14, 6))

plt.subplot(1, 2, 1)
plt.plot(t_num, x1, label = r"$x_1$ (t)", color = "red")
plt.plot(t_num, x2, label = r"$x_2$ (t)", color = "blue")
plt.plot(t_num, x3, label = r"$x_3$ (t)", color = "black")
plt.title("Laws of motion", color = "gold", fontweight = "bold")
plt.xlabel("Time t [s]")
plt.ylabel("Displacement [m]")
plt.legend()

ax = plt.subplot(1, 2, 2, projection = "3d")
ax.plot(x1, x2, x3, color = "black", label = "Trajectory")
ax.scatter(x1[0], x2[0], x3[0], color = "green", label = "Starting point", s = 50)
ax.scatter(x1[-1], x2[-1], x3[-1], color = "red", label = "End point", s = 50)
ax.set_title("3D Configuration space trajectory", color = "gold", fontweight = "bold")
ax.set_xlabel(r"$x_1$ (t) [m]")
ax.set_ylabel(r"$x_2$ (t) [m]")
ax.set_zlabel(r"$x_3$ (t) [m]")
ax.view_init(elev = 30, azim = 45) 
ax.legend()

plt.tight_layout()
plt.show()
```
